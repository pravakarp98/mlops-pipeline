```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
train = pd.read_csv("/home/pravakar/git_repo/ml_pipelines/data/train.csv")
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=train, x='Target', order=['Graduate', 'Dropout', 'Enrolled'], hue='Target')
plt.title('Distribution of target variable')
plt.xlabel('Academic Outcome')
plt.ylabel('Number of Students')
plt.savefig('target_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=train, x='Gender', hue='Target')
plt.title('Gender distribution by academic outcome')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Number of students')
plt.savefig('gender_vs_target.png')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.histplot(data=train, x='Age at enrollment', kde=True, bins=30)
plt.title('Distribution of Age at Enrollment')
plt.xlabel('Age')
plt.savefig('age_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=train, x='Target', y='Age at enrollment', order=['Graduate', 'Dropout', 'Enrolled'], hue='Target', orient='v')
plt.xlabel('Academic Outcome')
plt.ylabel('Age at enrollment')
plt.title('Age at Enrollment vs. Academic Outcome')
plt.savefig('age_vs_target_boxplot.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=train, x='Admission grade', kde=True)
plt.title('Distribution of Admission Grade')
plt.xlabel('Grade')
plt.savefig('admission_grade_distribution.png')
```

```python
numerical_cols = train.select_dtypes(include=['int64', 'float64']).columns
numerical_cols = numerical_cols.drop('id')

corr_matrix = train[numerical_cols].corr()

plt.figure(figsize = (18, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Correlation Heatmap of Numerical Features')
plt.savefig('correaltion_heatmap.png')
plt.show()
```
